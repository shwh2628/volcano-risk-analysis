```python
so2_datetime = pd.read_csv("/Users/shannonwhite/git/data/so22_ad_viz_plotval_data.csv",
                             parse_dates = ['Date'],
                             index_col = ['Date'])
```

```python
so2_datetime.sort_values(by="Daily Max 1-hour SO2 Concentration", ascending = False)
```

```python
fig, ax= plt.subplots()

# add the x-axis and the y-axis to the plot
ax.plot(so2_datetime.index.values, 
        so2_datetime['Daily Max 1-hour SO2 Concentration'], 
        color = 'red')

# rotate tick labels
plt.setp(ax.get_xticklabels(), rotation=45)

# set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Sulfur Dioxide Concentration",
       title="Sulfer Dioxide Concentration\nKilauea, Hawaii in 2018");
```

```python
# create the plot space upon which to plot the data
fig, ax= plt.subplots()

# add the x-axis and the y-axis to the plot
ax.bar(so2_datetime.index.values, 
        so2_datetime['Daily Max 1-hour SO2 Concentration'], 
        color = 'blue')

# rotate tick labels
plt.setp(ax.get_xticklabels(), rotation=45)

# set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Sulfur Dioxide Concentration",
       title="Sulfer Dioxide Concentration\nKilauea, Hawaii in 2018");
```
